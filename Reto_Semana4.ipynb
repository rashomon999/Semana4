#Ejercicio1

```
import matplotlib.pyplot as plt
import numpy as np


def fibonacci(n):
    fib_series = [0, 1]
    while len(fib_series) < n:
        fib_series.append(fib_series[-1] + fib_series[-2])
    return fib_series


n = int(input("Ingrese el número de términos de la serie de Fibonacci: "))


fib_series = fibonacci(n)



print("Primeros 10 resultados de la serie de Fibonacci:")
print(fib_series[:10])


plt.figure(figsize=(10, 6))
plt.plot(fib_series, marker='.')
plt.title(f'Serie de Fibonacci - Primeros {n} términos')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.grid(True)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9, 26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

media = np.mean(temperaturas)
print("Media:", media)

mediana = np.median(temperaturas)
print("Mediana:", mediana)

desviacion_estandar = np.std(temperaturas)
print("Desviación estándar:", desviacion_estandar)

temperaturasMensuales = temperaturas

mediasMensuales = [np.mean(temperaturas)for temperaturas in temperaturasMensuales]

plt.plot(range(20), mediasMensuales)
plt.xlabel('Mes')
plt.ylabel('Temperatura media (°C)')
plt.title('Temperatura media mensual')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Generar valores de n en el rango de 0 a 2π
n_values = np.linspace(0, 2 * np.pi, 100)

# Definición de las sucesiones
b_n = np.sin(n_values)
c_n = np.cos(n_values)
d_n = np.tan(n_values)

# Crear los gráficos
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(n_values, b_n)
plt.title('$b_{n} = \sin(n)$')
plt.xlabel('$n$')
plt.ylabel('$b_{n}$')

plt.subplot(2, 2, 2)
plt.plot(n_values, c_n)
plt.title('$c_{n} = \cos(n)$')
plt.xlabel('$n$')
plt.ylabel('$c_{n}$')

plt.subplot(2, 2, 3)
plt.plot(n_values, d_n)
plt.title('$d_{n} = \ tan(n)$')
plt.xlabel('$n$')
plt.ylabel('$d_{n}$')
plt.ylim(-10, 10)


plt.tight_layout()
plt.show()
```

#Ejercicio6

```
import matplotlib.pyplot as plt
import numpy as np

def calcular_pendiente(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)

def graficar_recta(x1, y1, x2, y2):
    pendiente = calcular_pendiente(x1, y1, x2, y2)
    x_vals = np.linspace(min(x1, x2) - 1, max(x1, x2) + 1, 400)
    y_vals = pendiente * (x_vals - x1) + y1

    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals)
    plt.text(x1, y1, f'({x1}, {y1})')
    plt.text(x2, y2, f'({x2}, {y2})')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gráfico de la recta')

    plt.grid(True)
    plt.show()

x1 = float(input("Ingrese x1: "))
y1 = float(input("Ingrese y1: "))
x2 = float(input("Ingrese x2: "))
y2 = float(input("Ingrese y2: "))

graficar_recta(x1, y1, x2, y2)
```

### Ejercicio 7

```
import numpy as np
import matplotlib.pyplot as plt

v0 = 20

def trayectoria(v0, theta):
        thetaRad = np.radians(theta)
        tTotal = 2 * v0 * np.sin(thetaRad) / 9.8
        t = np.linspace(0, tTotal, num=500)
        x = v0 * np.cos(thetaRad) * t
        y = v0 * np.sin(thetaRad) * t - 0.5 * 9.8 * t**2
        return x,y


thetas = [30, 45, 60, 75]

plt.figure(figsize=(10, 6))
for theta in thetas:
    x, y = trayectoria(v0, theta)
    plt.plot(x, y, label=f'{theta}°')

plt.title('Trayectoria Parabólica para Diferentes Ángulos de Lanzamiento')
plt.xlabel('Posición en el Eje X')
plt.ylabel('Posición en el Eje Y')
plt.grid(True)
plt.show()
```
